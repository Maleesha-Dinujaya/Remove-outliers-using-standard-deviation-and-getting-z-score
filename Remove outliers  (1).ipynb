```python
mean = df.Height.mean()
mean
```

```python
std_deviation = df.Height.std()
std_deviation
```

```python
low_outliers = mean-3*std_deviation
low_outliers
```

```python
High_outliers = mean+3*std_deviation
High_outliers
```

```python
df[(df.Height < low_outliers)|(df.Height > High_outliers)]
```
